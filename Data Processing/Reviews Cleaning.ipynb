```python
import datetime as dt
from datetime import datetime
import pandas as pd
import numpy as np
import uuid
```

```python
df = pd.DataFrame(pd.read_csv('../Data/reviews.csv'))
```

```python
df['date'] = uuid.uuid4()
df['id'] = uuid.uuid4()
```

```python
df
```

```python
for i in range(len(df)):
    df['id'][i] = uuid.uuid4()
    dt_object = datetime.fromtimestamp(df["timestamp"][i])
    df["date"][i] = dt_object.strftime("%F")
```

```python
df["text"] = "Blank"
```

```python
df["title"] = "N/A"
```

```python
cols = ['id', 'title', 'rating', 'text', 'movieId', 'timestamp', 'date', 'userId']
```

```python
df = df[cols]
```

```python
df
```

```python
df.to_csv('../Data/reviews.csv', index=False)
```

```python
dffilm = pd.DataFrame(pd.read_csv('../Data/movieinfo.csv'))
dfrating = pd.DataFrame(pd.read_csv('../Data/ratings.csv'))
```

```python
dfrating.head()
```

```python
mean_movies_rat = dfrating.groupby(['movieId']).agg({'rating': np.mean}).reset_index()
```

```python
mean_movies_rat.head()
```

```python
df3 = pd.merge(dffilm, mean_movies_rat, left_on='movieId', right_on='movieId', how='left')
```

```python
df3
```

```python
df3.to_csv('../Data/movieinfo.csv', index=False)
```
